```python
df["Sales"] = (
    df["Sales"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["Sales"] = df["Sales"].astype(float)
```

```python
df["Profit"] = (
    df["Profit"]
    .str.replace("-$", "-", regex=False)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["Profit"] = df["Profit"].astype(float)
```

```python
df[["Sales", "Profit"]].dtypes
```

```python
df["Discount"] = (
    df["Discount"]
    .str.replace("%", "", regex=False)
    .astype(float)
)
```

```python
df["Profit Ratio"] = (
    df["Profit Ratio"]
    .str.replace("%", "", regex=False)
    .astype(float)
)
```
